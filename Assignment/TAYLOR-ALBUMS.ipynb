I AM TRYING TO MAKE A CORPUS FROM TAYLOW SWIFT'S 2 ALBUM, The "Taylor Swift" and "Evermore" albums.

```python
import os
import pandas as pd
import spacy

# Load the spaCy model
nlp = spacy.load('en_core_web_sm')

# Define the path to the folder
swiftie_folder = "../Swiftie"
def process_csv(file_path):
    df = pd.read_csv(file_path)
    text_data = df['lyric'].tolist()  
    return text_data
```

```python
taylor_swift_csv = "../Swiftie/01-taylor_swift.csv"
evermore_csv = "../Swiftie/09-evermore_deluxe_version.csv"
taylor_swift_documents = process_csv(taylor_swift_csv)
evermore_documents = process_csv(evermore_csv)
```

```python
texts = []
for document in taylor_swift_documents:
    text = []
    doc = nlp(document)
    for w in doc:
        if not w.is_stop and not w.is_punct and not w.like_num:
            text.append(w.lemma_) # Appends the lemmatized form of the word to the text list. Lemmatization involves reducing words to their base or root form.
    texts.append(text)
#texts is a mini-corpus specifically for toxic algal bloom
print(texts)
```

```python
texts1 = []
for document in evermore_documents:
    text = []
    doc = nlp(document)
    for w in doc:
        if not w.is_stop and not w.is_punct and not w.like_num:
            text.append(w.lemma_) # Appends the lemmatized form of the word to the text list. Lemmatization involves reducing words to their base or root form.
    texts1.append(text)
#texts is a mini-corpus specifically for toxic algal bloom
print(texts1)
```

```python
#creating a BOW representation of the mini-corpus
from gensim import corpora
dictionary = corpora.Dictionary(texts)
print(dictionary.token2id)
```

```python
#creating a BOW representation of the mini-corpus
from gensim import corpora
dictionary1 = corpora.Dictionary(texts1)
print(dictionary1.token2id)
```

```python
corpus = [dictionary.doc2bow(text) for text in texts]
corpus
```

```python
corpus1 = [dictionary.doc2bow(text) for text in texts1]
corpus1
```

```python
#storing your generated corpus

corpora.MmCorpus.serialize('Taylor_Swift_album.mm', corpus)
corpora.MmCorpus.serialize('Evermore_album.mm', corpus1)
```
